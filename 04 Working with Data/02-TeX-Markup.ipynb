# TeX Markup in Matplotlib

Matplotlib supports a subset of LaTeX for writing math symbols, equations, fractions, square roots, and more.

We use **`r'$ ... $'`** to wrap the equation and render it using math text.

ðŸ”— Reference: [MathText â€” Matplotlib Docs](https://matplotlib.org/stable/tutorials/text/mathtext.html)


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for line (can be from earlier)
x_1 = np.linspace(0, 5, 50)
y_1 = x_1 ** 2

fig_5 = plt.figure(figsize=(5, 4), dpi=100)
axes_5 = fig_5.add_axes([0.1, 0.1, 0.9, 0.9])

# Greek letters
axes_5.text(0, 23, r'$\alpha \beta \sigma \omega \epsilon \mu \pi \theta \lambda$', fontsize=18)

# Subscripts, superscripts, sum, fractions, binomial
axes_5.text(0, 18, r'$\delta_i \gamma^{ij} \sum_{i=0}^\infty x_i \frac{3}{4} \binom{3}{4}$', fontsize=18)

# Fraction inside a fraction, square root, trig, cube root
axes_5.text(0, 13, r'$\frac{8 - \frac{x}{5}}{8} \sqrt{9} \sin(\pi) \sqrt[3]{8} \acute a \div$', fontsize=18)

# Decorations and limits
axes_5.text(0, 8, r'$\bar a \hat a \tilde a \vec a \overline {a} \lim_{x \to 2} f(x) = 5$', fontsize=18)

# Comparison symbols
axes_5.text(0, 3, r'$\geq \leq \ne$', fontsize=18)

# Optional line plot for background context
axes_5.plot(x_1, y_1)
plt.show()
```

## Summary

- Use `r'$...$'` to render math formulas
- Works great for annotations, text, legends, and titles
- Keep it clean using raw strings (`r''`) to avoid escape issues
