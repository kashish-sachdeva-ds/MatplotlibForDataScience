#  Histogram of Sum of Two Dice Rolls

In this notebook, we simulate rolling **two 6-sided dice** 5000 times and then plot a **histogram of their sum** using Matplotlib.

---

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Roll Two Dice 5000 Times

We use NumPy's `randint()` to simulate 5000 rolls for each die.

Then we add the values to get the sum of both dice.


```python
# Roll 2 six-sided dice and get their sum
arr_1 = np.random.randint(1, 7, 5000)  # Dice 1
arr_2 = np.random.randint(1, 7, 5000)  # Dice 2
arr_3 = arr_1 + arr_2                  # Sum of both dice
```

## Step 2: Plot the Histogram

We now plot the histogram using Matplotlib with the following settings:

- `bins=11`: Because the possible sums range from 2 to 12
- `density=True`: Shows normalized frequency (area under curve = 1)
- `stacked=True`: Stack bars if needed (used for multi-dataset comparison)
- `color='orange'`: Makes the bars orange

Other optional parameters (not used here, but mentioned in comments):

- `range=(2,12)`: Set the range of bins
- `cumulative=True`: Shows cumulative distribution
- `histtype='step'`: Makes line-like histogram
- `orientation='horizontal'`: Horizontal bars instead of vertical


```python
# Plot the histogram of the sums
plt.figure(figsize=(8,5))
plt.hist(arr_3, bins=11, density=True, stacked=True, color='orange', edgecolor='black')

# Add labels and title
plt.title('Histogram of Sum of Two Dice Rolls (5000 trials)')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Normalized Frequency')
plt.grid(True)
plt.show()
```

## Observation

- The most frequent sum is **7**, as it has the most combinations (like 1+6, 2+5, 3+4, etc.)
- The histogram has a **triangular shape**, peaking at 7 and tapering off toward 2 and 12.
- This is a common pattern when summing uniform distributions like dice rolls.
