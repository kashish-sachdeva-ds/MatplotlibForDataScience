# Using `plt.subplots()` for Multiple Axes

In this notebook, we'll learn:
- How to use `plt.subplots()` to auto-create multiple axes
- How to arrange plots in rows and columns
- How to access individual axes and customize them


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x_1 = np.linspace(0, 5, 50)
y_1 = x_1 ** 2
```

## Step 1: Create Subplots Automatically

`plt.subplots()` creates both:
- A **figure object**
- And a **grid of axes objects**

We can control the layout using `nrows` and `ncols`.


```python
# 1 row, 3 columns â†’ gives us 3 subplots
fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3)

# Add space between the plots
plt.tight_layout()
```

## Step 2: Access and Customize Each Plot

You can use indexing (like `axes_2[1]`) to work with individual plots.


```python
# Customize the middle subplot (index 1)
axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1, y_1)

plt.show()
```

## `Now` Put It All Together (Run This At Once)

```python

fig_2, axes_2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)

plt.tight_layout()

axes_2[1].set_title('Plot 2')
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('X Squared')
axes_2[1].plot(x_1,y_1)
```

## Summary

- `plt.subplots()` is the easiest way to create a grid of plots
- You get a list (or 2D array) of axes objects to work with
- Use indexing like `axes[0]`, `axes[1]` to access and customize each one
- Use `plt.tight_layout()` to prevent overlaps between subplots
