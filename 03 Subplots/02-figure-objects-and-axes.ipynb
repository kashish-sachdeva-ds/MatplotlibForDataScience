# Using Figure Objects and Axes in Matplotlib

In this notebook, you will learn:
- What is a `Figure` object in Matplotlib
- How to manually add `Axes` using `.add_axes()`
- How to plot multiple lines, add labels, title, legend
- How to nest an axis inside another (inset plot)
- How to insert custom text into a plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x_1 = np.linspace(0, 5, 50)
y_1 = x_1 ** 2
```

## Step 1: Create a Figure Canvas

A **Figure** is the outermost container — like a blank whiteboard.  
We define its size using `figsize=(width, height)` and DPI (dots per inch).

```python
fig_1 = plt.figure(figsize=(5, 4), dpi=100)
```

## Step 2: Add Axes to the Figure

Use `.add_axes([left, bottom, width, height])`  
These values are in percentage (from 0 to 1) of the figure space.

```python
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])  # Main plot area
```

## Step 3: Set Labels and Title

```python
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')
```

##  Step 4: Plot Multiple Lines and Add Legend

We use `.plot(x, y, label='...')`  
The `label` is used in the legend.

```python
axes_1.plot(x_1, y_1, label='x / x²')
axes_1.plot(y_1, x_1, label='x² / x')
axes_1.legend(loc=0)  # Best location
```

## Step 5: Add an Inset Plot Inside the Main Plot

You can place another small plot within the main one.

```python
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Small Days Squared Chart')
axes_2.plot(x_1, y_1, 'r')
```

## Step 6: Add Custom Text to the Plot

```python
axes_2.text(0, 40, 'Message')
```

---
## `Now` Put It All Together (Run This At Once)

If you're running cells step-by-step, variables like `fig`, `axes` may get lost from memory.  
So for consistent results, always run the full plot in one go.

```python
fig_1 = plt.figure(figsize=(5, 4), dpi=100)

axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])

axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days Squared')
axes_1.set_title('Days Squared Chart')

axes_1.plot(x_1, y_1, label = 'x / x²')
axes_1.plot(y_1, x_1, label = 'x² / x')
# Add the optional legend with a location number (best: 0, 
# upper right: 1, upper left: 2, lower left: 3, lower right: 4,
# https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.legend.html)
# or supply a tuple of x & y from lower left
axes_1.legend(loc = 0)

# You can create axis inside of others
axes_2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes_2.set_xlabel('Days')
axes_2.set_ylabel('Days Squared')
axes_2.set_title('Small Days Squared Chart')
axes_2.plot(x_1, y_1,'r')

# Add text to plot from central point of 0,0
axes_2.text(0, 40, 'Message')
```

## Summary

- `plt.figure()` creates the figure canvas
- `.add_axes()` gives full control over where your plots go
- You can insert inset plots and even overlay text
- Useful for detailed dashboards and multiple views
