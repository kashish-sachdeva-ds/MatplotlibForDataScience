# Appearance Customization in Matplotlib

In this notebook, we explore:
- Line color, opacity, and width
- Line styles and marker styles
- Marker colors and borders
- Grid styling and background color


````python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x_1 = np.linspace(0, 5, 50)
y_1 = x_1 ** 2
````

## Step 1: Create the Figure and Axes


````python
fig_3 = plt.figure(figsize=(6, 4))
axes_3 = fig_3.add_axes([0, 0, 1, 1])
````

## Step 2: Customize Plot Appearance

We apply:
- `color='navy'` → dark blue line
- `alpha=0.75` → 75% opacity
- `lw=2` → line width = 2
- `ls='-.'` → dash-dot line
- `marker='o'` → circle marker
- `markersize=7` → size of each marker
- `markerfacecolor='y'` → fill color = yellow
- `markeredgecolor='y'` → border color = yellow
- `markeredgewidth=4` → border thickness

````python
axes_3.plot(
    x_1, y_1,
    color='navy', alpha=0.75, lw=2, ls='-.',
    marker='o', markersize=7,
    markerfacecolor='y',
    markeredgecolor='y', markeredgewidth=4
)
````

## Step 3: Set Axis Range and Background

````python
axes_3.set_xlim([0, 3])
axes_3.set_ylim([0, 8])
axes_3.set_facecolor('#FAEBD7')  # light antique white
````

## Step 4: Add a Styled Grid

````python
# Add custom dashed grid
axes_3.grid(True, color='.6', dashes=(5, 2, 1, 2))
plt.show()
````

## `Now` Put It All Together (Run This At Once)

````python
fig_3 = plt.figure(figsize=(6,4))
axes_3 = fig_3.add_axes([0,0,1,1])

axes_3.plot(
    x_1, y_1,
    color='navy', alpha=0.75, lw=2, ls='-.',
    marker='o', markersize=7,
    markerfacecolor='y',
    markeredgecolor='y', markeredgewidth=4
)

axes_3.set_xlim([0,3])
axes_3.set_ylim([0,8])
axes_3.grid(True, color='0.6', dashes=(5, 2, 1, 2))

axes_3.set_facecolor('#FAEBD7')
````

## Summary

- Use keyword arguments to customize lines, markers, and colors
- Add `grid()` and `set_facecolor()` to improve visual readability
- Always finish with `plt.show()` to render your styled plot
