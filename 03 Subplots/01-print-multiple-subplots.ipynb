# Print Multiple Plots with `plt.subplot()`

In this notebook, we’ll learn how to:
- Display multiple plots in the same figure
- Use `plt.subplot(rows, cols, index)` to place plots


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Create Sample Data

```python
x1 = np.linspace(0, 5, 50)
y1 = x1 ** 2
```

## Plot Two Subplots Side-by-Side
- One will be red
- One will be blue and reversed


```python
plt.subplot(1, 2, 1)  # 1 row, 2 cols, 1st plot
plt.plot(x1, y1, 'r')  # red line

plt.subplot(1, 2, 2)  # 1 row, 2 cols, 2nd plot
plt.plot(y1, x1, 'b')  # blue line, reversed

plt.tight_layout()
plt.show()
```

## Summary

- `plt.subplot(1, 2, 1)` = First plot in a 1x2 layout
- `plt.subplot(1, 2, 2)` = Second plot beside it
- Great for visual comparison!
