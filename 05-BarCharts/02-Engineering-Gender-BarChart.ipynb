# Gender Distribution in Engineering Fields â€“ Grouped Bar Chart

This notebook visualizes the **percentage of male and female students** in different branches of engineering using a **grouped bar chart**.

It helps to compare how representation varies by gender across fields like:
- Aerospace
- Chemical
- Civil
- Electrical
- Mechanical

```python
import numpy as np
import matplotlib.pyplot as plt

# Percentage of male and female engineers by discipline
m_eng = (76, 85, 86, 88, 93)
f_eng = (24, 15, 14, 12, 7)

# Get evenly spaced values for bars
spc = np.arange(5)

plt.figure(figsize=(8,5))

# Plot male and female bars side by side
plt.bar(spc, m_eng, width=0.45, label='Male', edgecolor='black', color='skyblue')
plt.bar(spc + 0.45, f_eng, width=0.45, label='Female', edgecolor='black', color='salmon')

# Set x-axis labels centered between grouped bars
plt.xticks(spc + 0.45 / 2, ('Aero', 'Chem', 'Civil', 'Elec', 'Mech'))

# Add labels, legend, grid
plt.title("Gender Distribution in Engineering Fields")
plt.ylabel("Percentage (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
