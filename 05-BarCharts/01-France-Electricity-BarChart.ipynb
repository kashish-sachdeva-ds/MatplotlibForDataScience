In this notebook, we analyze how France generates its electricity using various energy sources.

We use a **horizontal bar chart** with **error bars** to show:
- the percentage contribution of each source
- the variance or fluctuation in the data (as x-error bars)

---


```python
import numpy as np
import matplotlib.pyplot as plt

# Electricity sources and usage %
x = ['Nuclear', 'Hydro', 'Coal', 'Gas', 'Solar', 'Wind', 'Other']
per_1 = [71, 10, 3, 7, 2, 4, 3]

# Variance in percentage
variance = [8, 3, 1, 3, 1, 2, 1]

# Create horizontal bar chart with error bars
plt.figure(figsize=(8, 5))
plt.barh(x, per_1, xerr=variance, color='teal', ecolor='black', edgecolor='black')
plt.title("France's Electricity Sources (with Variance)")
plt.xlabel("Percentage (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
