# Teachers by Category and Gender â€“ Stacked Bar Chart

This notebook presents the number of **male and female teachers** across different teaching categories using a **stacked bar chart**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Teacher types and counts
t_type = ['Kind', 'Elem', 'Sec', 'Spec']
m_teach = np.array([2, 20, 44, 14])
f_teach = np.array([98, 80, 56, 86])
ind = [x for x, _ in enumerate(t_type)]

plt.figure(figsize=(8,5))

# Stacked bar chart
plt.bar(ind, m_teach, width=0.45, label='Male', bottom=f_teach, color='navy')
plt.bar(ind, f_teach, width=0.45, label='Female', color='orange')

# Labeling and legends
plt.xticks(ind, t_type)
plt.title("Gender Distribution of Teachers by Category")
plt.ylabel("Number of Teachers")
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
