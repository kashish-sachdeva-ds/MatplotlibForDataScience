# Pokémon Type Distribution – Pie Chart

In this notebook, we visualize the **distribution of Pokémon by type** using a **custom pie chart**. The chart shows how many Pokémon belong to each type, using:

- Random dark color palette
- Highlighted top slice
- Shadow effect
- Legends outside the pie

---


```python
import matplotlib.pyplot as plt
import random

# Set figure size and resolution
fig_6 = plt.figure(figsize=(8, 5), dpi=100)
axes_6 = fig_6.add_axes([0.1, 0.1, 0.9, 0.9])

# Types and their respective Pokémon counts
types = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
         'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy',
         'Ghost', 'Ice']
poke_num = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 40]

# Generate random dark colors
colors = []
for _ in range(18):
    rgb = (random.uniform(0, 0.5), random.uniform(0, 0.5), random.uniform(0, 0.5))
    colors.append(rgb)

# Explode (highlight) the largest slice
explode = [0] * 18
explode[0] = 0.2  # Water type

# Create the pie chart
wedges, texts, autotexts = plt.pie(
    poke_num,
    explode=explode,
    labels=types,
    colors=colors,
    autopct='%1.0f%%',
    shadow=True,
    startangle=140,
    textprops=dict(color="w")
)

# Add legend outside the pie
plt.legend(wedges, types, loc='right', bbox_to_anchor=(1, 0, 0.5, 1))

plt.title("Distribution of Pokémon by Type")
plt.show()
```
