# Customize Line Style in Matplotlib

In this notebook, we’ll learn how to:
- Change line color and thickness
- Add markers (like dots or squares)
- Use different line styles (dashed, dotted, etc.)


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Step 1: Sample Data

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 20, 25, 30])
```

## Step 2: Try Different Line Styles

We’ll try:
- `color='red'` to change line color
- `linewidth=2` for thickness
- `linestyle='--'` for dashed lines
- `marker='o'` to show dots on points


```python
plt.plot(x, y, color='red', linewidth=2, linestyle='--', marker='o')

plt.title("Customized Line Style")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

## Summary

- Use `color`, `linewidth`, and `linestyle` to style the line
- Use `marker` to show data points
- You can combine all styles in one `plt.plot()` call
