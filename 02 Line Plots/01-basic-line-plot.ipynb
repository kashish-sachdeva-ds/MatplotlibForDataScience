# Basic Line Plot using Matplotlib

In this notebook, we will learn how to:
- Create a simple line plot using `plt.plot()`
- Add a title, x-axis and y-axis labels
- Display the plot inside the notebook


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Step 1: Create Sample Data

We'll create two simple NumPy arrays:
- `x` for horizontal values
- `y` for vertical values


```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 20, 25, 30, 40])
```

## Step 2: Plot the Line Graph

We'll use `plt.plot()` to draw a line using the values in `x` and `y`.


```python
plt.plot(x, y)  # Draws a line chart with x and y

# Add title and labels
plt.title("Basic Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Show the graph : # If not in Jupyter Notebook use this
plt.show()
```

## Example:

- Create numpy arrays : `x` & `y` (10 Floats from 0 to 5)

```python
x = np.linspace(0,5,10)
y = x**2
```

```python
plt.plot(x, y)

plt.title('Days Squared Chart')

plt.xlabel('Days')
plt.ylabel('Days Squared')
```

---
> Note: Your x-axis and y-axis points must have equal length.
For example, if you have 5 x-values, then you must also have 5 y-values — only then the plot will work properly.
---

## Summary

- `plt.plot()` draws a simple line chart
- Always use `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` to describe your plot
- `plt.show()` is required to actually display the graph
